```python
import pandas as pd
df=pd.read_csv('/home/saravananmsk1536/Bigdata/Python/Input/song_duration.csv')
print(df)
```

```python
result=df.groupby('user_id').agg({'listen_duration': 'sum', 'song_id': 'nunique'}).reset_index()
result.columns = ['user_id', 'total_listen_duration', 'unique_song_count']
print(result)
```
