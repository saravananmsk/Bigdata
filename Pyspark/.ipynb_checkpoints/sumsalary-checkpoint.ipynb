```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,sum
sc = SparkSession.builder.appName('"textflile"').getOrCreate()
data=[("msk", "test1@gmail.com", 5000), ("Ram", "test2@gmail.com", 7000), ("Remya", "test3@gmail.com", 8000)]
schema="name string, email string, salary int"
df=sc.createDataFrame(data, schema)
df.show()
dfsum=df.agg(sum(col("salary")).alias("total_salary"))
dfsum.show()
```
